```python
import pandas as pd 
import json 
import copy
import re 
from collections import Counter, defaultdict

pd.set_option('display.max_columns',100)
pd.set_option('display.max_colwidth',500)


path1 = '../cache/bert-large-uncased/lm_diagnostic_extended/sgpl/exp_data_results_anchor_type_Coordinate_remove_Y_PUNC_FULL_concate_or_single_max_anchor_num_5_anchor_scorer_probAvg_wnp_True_cpt_False.LM_DIAGNOSTIC_EXTENDED.csv'

path2 = '../cache/bert-large-uncased/lm_diagnostic_extended/sgpl/exp_data_results_anchor_type_Coordinate_remove_Y_PUNC_FULL_concate_or_single_max_anchor_num_5_anchor_scorer_probAvg_wnp_True_cpt_True.LM_DIAGNOSTIC_EXTENDED.csv'


df1 = pd.read_csv(path1)
df1.subj_anchors = df1.subj_anchors.apply(lambda x: eval(x))


df2 = pd.read_csv(path2)
df2.subj_anchors = df2.subj_anchors.apply(lambda x: eval(x))
```

```python

df_new = pd.merge(df1, df2, on = ['sub_label', 'obj_label']
                                  , suffixes=("_x", "_y"))
df_new['shared_anchors'] = df_new[['subj_anchors_x', 'subj_anchors_y']].apply(lambda x: set(x[0]).intersection(set(x[1])), axis=1)
df_new['unique_anchors_1'] = df_new[['subj_anchors_x', 'subj_anchors_y']].apply(lambda x: set(x[0]) - set(x[1]), axis=1)
df_new['unique_anchors_2'] = df_new[['subj_anchors_x', 'subj_anchors_y']].apply(lambda x: set(x[1]) - set(x[0]), axis=1)
df_new['shared_anchors_ratio_1'] = df_new[['shared_anchors', 'subj_anchors_x']].apply(lambda x:  len(x[0])/len(x[1]) , axis=1)
df_new['shared_anchors_ratio_2'] = df_new[['shared_anchors', 'subj_anchors_y']].apply(lambda x:  len(x[0])/len(x[1]) , axis=1)
print("shared anchors in *_x: ", df_new['shared_anchors_ratio_1'].mean())
print("shared anchors in *_y: ", df_new['shared_anchors_ratio_2'].mean())
#     return df_new 

# df_new[['sub_label', 'shared_anchors']].head()
# df_new.columns
# df_new = compare_anchors(df1, df2)
# df_new[['sub_label', 'unique_anchors_1', 'unique_anchors_2']]

count =0 
for x in df_new[['sub_label', 'obj_label', 'unique_anchors_1', 'unique_anchors_2']].to_dict(orient='record'):
    if len(x['unique_anchors_1']) >0: 
        print(x)
        count +=1
print(f"count: {count}")
#     .items():
#     if len(v)>0:
#         print(k, v)

```
