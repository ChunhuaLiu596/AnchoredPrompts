```python

# from significance_test import significance_test_multiple_datasets
import pandas as pd 
from scipy import stats 
from statsmodels.stats.multitest import multipletests
from statsmodels.stats.contingency_tables import mcnemar

def significance_test_multiple_datasets_1(dataset_to_df, metrics, min_pvalue =0.05, display_res=False):
    df_ttest_rel = []
    for dataset,df in dataset_to_df.items(): 
        # metrics = [['p1_def_sap', 'p1_def_dap'], ['p1_lsp_sap','p1_lsp_dap'], ['p1_def_dap', 'p1_lsp_dap']]
        for metric in metrics: 
            statistic, pvalue = stats.ttest_rel(df[metric[0]], df[metric[1]])
            reject_np = True if pvalue < min_pvalue else False 
            df_ttest_rel.append({'metric': ' .vs '.join(metric), 'dataset': dataset, 'statistic': statistic, 'pvalue': pvalue, 'significantly different': reject_np})
    #         print(dataset, statistic, pvalue, reject_np)

    df_ttest_rel = pd.DataFrame(df_ttest_rel)
    # display(df_ttest_rel.sort_values(by=['metric', 'significantly different']))
    if display_res: 
        for name, group in df_ttest_rel.groupby('metric'):
            display(group.sort_values('significantly different'))
    return df_ttest_rel

    
query_datasets = list(dataset_to_localpath.keys())
dataset_to_df_test = {k:v for k,v in dataset_to_df.items() if k in query_datasets}
significance_test = defaultdict(dict)
significance_test_adjusted = defaultdict(dict)

significance_test_score = defaultdict(dict)
significance_test_adjusted_score = defaultdict(dict)


metrics =[
          ['p10_def_sap', 'p10_lsp_sap'], 
          ['p10_def_sap', 'p10_def_dap'],
          ['p10_lsp_sap', 'p10_lsp_dap'],
          ['mrr_def_sap', 'mrr_lsp_sap'],
          ['mrr_def_sap', 'mrr_def_dap'], 
          ['mrr_lsp_sap', 'mrr_lsp_dap']
]

min_pvalue = 0.05 
for metric_pair in metrics:
    df_ttest_rel=significance_test_multiple_datasets_1(dataset_to_df_test, [metric_pair], min_pvalue=0.05, display_res=False)
    
    significance_test[df_ttest_rel['metric'][0]] = dict(zip(df_ttest_rel['dataset'], df_ttest_rel['significantly different']))
    significance_test_score[df_ttest_rel['metric'][0]] = dict(zip(df_ttest_rel['dataset'], df_ttest_rel['pvalue']))

    
    p_adjusted = multipletests(df_ttest_rel['pvalue'], method='holm')[1] # holm fdr_bh
#     p_adjusted = multipletests(df_ttest_rel['pvalue'], method='bonferroni')[1] # bonferroni: dependent datasets
    
    p_adjusted_res = [True if x< min_pvalue else False  for x in p_adjusted ]
    
    significance_test_adjusted[df_ttest_rel['metric'][0]] = dict(zip(df_ttest_rel['dataset'], p_adjusted_res))
    significance_test_adjusted_score[df_ttest_rel['metric'][0]] = dict(zip(df_ttest_rel['dataset'], p_adjusted))
        
    #display(significance_test)
    #display(p_adjusted_dict)
    
print("Before bonferroni correction" )
display(pd.DataFrame(significance_test).T)
display(pd.DataFrame(significance_test_score).T)


print("After bonferroni correction" )
display(pd.DataFrame(significance_test_adjusted).T)

display(pd.DataFrame(significance_test_adjusted_score).T)

# print(json.dumps(significance_test, indent=4))
# print(json.dumps(significance_test_adjusted, indent=4))
```

```python

from sklearn.metrics import confusion_matrix
from statsmodels.stats.contingency_tables import mcnemar

def mcnemar_test(list_1, list_2):
    """
    Computes the McNemar test for two lists of paired data.

    Args:
        list_1 (list): List of binary outcomes for group 1
        list_2 (list): List of binary outcomes for group 2

    Returns:
        Tuple containing chi-square statistic and p-value
    """
    tn, fp, fn, tp = confusion_matrix(list_1, list_2).ravel()
    return mcnemar([[tn, fp], [fn, tp]])




result = mcnemar_test(dataset_to_df_test['BLESS']['p10_def_sap'], dataset_to_df_test['BLESS']['p10_def_dap'])
print(result)
```
